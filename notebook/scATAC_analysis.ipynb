```python
# Import necessary libraries
import scanpy as sc
import anndata as ad
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import scrublet as scr
```

```python
# Define dataset URL and filename (using a PBMC dataset as an example)
dataset_url = "https://cf.10xgenomics.com/samples/cell-atac/1.1.0/atac_pbmc_10k_v1/atac_pbmc_10k_v1_filtered_peak_bc_matrix.h5"
dataset_filename = "atac_pbmc_10k_filtered_peak_bc_matrix.h5"
if not os.path.exists(dataset_filename):
    print(f"Downloading dataset from {dataset_url}...")
    os.system(f"wget -O {dataset_filename} {dataset_url}")

# Load the dataset from the provided URL or use a demo dataset from Scanpy
adata = sc.read_10x_h5(dataset_filename)

adata = sc.datasets.pbmc3k()
print(f"Loaded dataset with {adata.n_obs} cells and {adata.n_vars} genes.")
```

# Step 1: Basic Quality Control and Filtering

```python
# Calculate mitochondrial gene content
mito_genes = adata.var_names.str.startswith('MT-')
adata.obs['percent_mt'] = np.sum(adata[:, mito_genes].X, axis=1) / np.sum(adata.X, axis=1)

# Calculate total counts per cell and number of genes expressed per cell
adata.obs['total_counts'] = adata.X.sum(axis=1)
adata.obs['n_genes'] = (adata.X > 0).sum(axis=1)
adata.obs['log_counts'] = np.log1p(adata.obs['total_counts'])

# Plot QC metrics
sc.pl.violin(adata, ['n_genes', 'total_counts', 'percent_mt'], jitter=0.4, multi_panel=True)
sc.pl.scatter(adata, x='total_counts', y='n_genes')
sc.pl.scatter(adata, x='total_counts', y='percent_mt')

# Filter cells based on QC metrics 
adata = adata[(adata.obs['n_genes'] > 200) & (adata.obs['n_genes'] < 2500) & (adata.obs['percent_mt'] < 5), :]
print(f"After quality control filtering: {adata.n_obs} cells remain.")
```

# Step 2: Doublet Detection using Scrublet

```python
scrub = scr.Scrublet(adata.X)  # Initialize Scrublet object
doublet_scores, predicted_doublets = scrub.scrub_doublets()
adata.obs['doublet_score'] = doublet_scores  
adata.obs['predicted_doublet'] = predicted_doublets

# Plot the histogram of doublet scores
scrub.plot_histogram()
plt.show()

# Filter out predicted doublets
adata = adata[~adata.obs['predicted_doublet'], :]
print(f"After doublet removal: {adata.n_obs} cells remain.")
```

# Step 3: Normalization and Log Transformation

```python
# Normalize total counts to 10,000 per cell and log-transform the data
sc.pp.normalize_total(adata, target_sum=1e4)  
sc.pp.log1p(adata)
```

# Step 4: Highly Variable Genes (HVGs) Identification

```python
# Identify highly variable genes using Scanpy's built-in function
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
sc.pl.highly_variable_genes(adata)  
adata = adata[:, adata.var.highly_variable]  
```

# Step 5: Scale and Center the Data

```python
sc.pp.scale(adata, max_value=10)  # Scale each gene to unit variance and center around 0
```

# Step 6: Dimensionality Reduction using PCA and NMF

```python
# Perform PCA for dimensionality reduction
sc.tl.pca(adata, svd_solver='arpack')
sc.pl.pca_variance_ratio(adata, log=True)  # Plot explained variance of PCA components
```

# Optionally, use Non-negative Matrix Factorization (NMF) for dimensionality reduction

```python
from sklearn.decomposition import NMF

# Check for negative values and adjust if necessary
if (adata.X < 0).any():
    print("Negative values detected. Adjusting to make all values non-negative...")
    adata.X = adata.X - np.min(adata.X) + 1  # Shift values to ensure all are positive

# Apply NMF
nmf_model = NMF(n_components=50, random_state=0)
adata.obsm['X_nmf'] = nmf_model.fit_transform(adata.X)  # Use NMF to reduce dimensions
```

# Step 7: Neighbors and UMAP computation based on NMF

```python
sc.pp.neighbors(adata, n_neighbors=15, use_rep='X_nmf')  # Compute nearest neighbors using NMF-reduced data
sc.tl.umap(adata)  # Compute UMAP embeddings
sc.pl.umap(adata, color=['total_counts', 'n_genes', 'percent_mt'])  # UMAP plot colored by QC metrics
```

# Step 8: Clustering using the Leiden Algorithm

```python
sc.tl.leiden(adata, resolution=0.8)  # Cluster cells using Leiden algorithm at a higher resolution
sc.pl.umap(adata, color='leiden')  # UMAP plot colored by clusters

# Additional clustering visualization with t-SNE
sc.tl.tsne(adata, n_pcs=40)
sc.pl.tsne(adata, color='leiden')  # t-SNE plot colored by clusters
```

# Step 9: Advanced Differential Expression Analysis

```python
# Identify marker genes using the Wilcoxon rank-sum test
sc.tl.rank_genes_groups(adata, 'leiden', method='wilcoxon')
sc.pl.rank_genes_groups(adata, n_genes=25, sharey=False)  # Plot top 25 marker genes for each cluster

# Additional visualization for marker genes using dotplot and matrixplot
sc.pl.rank_genes_groups_dotplot(adata, n_genes=4)  # Dotplot of top 4 marker genes
sc.pl.rank_genes_groups_matrixplot(adata, n_genes=4)  # Matrixplot of top 4 marker genes
```

# Step 10: Save Processed and Improved Data

```python
output_file_improved = "pbmc3k_improved_data_v2.h5ad"
adata.write(output_file_improved)
print(f"Enhanced dataset saved to {output_file_improved}.")
```
